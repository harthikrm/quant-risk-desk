# 03 Correlation Dynamics
Investigating the evaporation of portfolio diversification during Black Swan events (e.g. COVID 2020).

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
import os; from dotenv import load_dotenv

load_dotenv("../.env")
engine = create_engine(f"postgresql://{os.getenv("POSTGRES_USER")}:{os.getenv("POSTGRES_PASSWORD")}@localhost:5433/risk_db")
```

```
corr_df = pd.read_sql("SELECT * FROM mart_correlation WHERE etf_1 = 'XLK' AND etf_2 = 'XLF'", engine)
corr_df["date"] = pd.to_datetime(corr_df["date"])

plt.figure(figsize=(12,5))
plt.plot(corr_df["date"], corr_df["correlation_63d"], color="purple")
plt.axvspan(pd.to_datetime("2020-02-20"), pd.to_datetime("2020-04-01"), color="red", alpha=0.3, label="COVID Crash")
plt.title("Tech & Financials Rolling Correlation (Spikes to 1.0 in Crash)")
plt.legend()
plt.show()
```
