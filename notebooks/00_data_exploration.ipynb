# 00 Data Exploration & Base Statistics
Initial profile of the raw daily return series to verify data hygiene.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
import os
from dotenv import load_dotenv

plt.style.use("seaborn-v0_8-darkgrid")
load_dotenv("../.env")

DB_USER = os.getenv("POSTGRES_USER", "postgres")
DB_PASS = os.getenv("POSTGRES_PASSWORD", "postgres")
DB_HOST = os.getenv("POSTGRES_HOST", "localhost")
DB_PORT = "5433"
DB_NAME = os.getenv("POSTGRES_DB", "risk_db")
engine = create_engine(f"postgresql://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_NAME}")
```

```
returns_df = pd.read_sql("SELECT * FROM int_daily_returns", engine)
returns_df["date"] = pd.to_datetime(returns_df["date"])
returns_df.head()
```

```
# Plot cumulative returns for SPY and XLK
plt.figure(figsize=(12,6))
for ticker in ["SPY", "XLK"]:
    ticker_data = returns_df[returns_df["ticker"] == ticker].sort_values("date")
    cumulative = np.exp(ticker_data["log_return"].cumsum()) - 1
    plt.plot(ticker_data["date"], cumulative, label=ticker)
plt.title("Cumulative Returns: SPY vs Tech (XLK)")
plt.legend()
plt.show()
```
