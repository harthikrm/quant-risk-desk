# 02 Macro Regime Analysis
How do different sectors perform across specific macro shifts (Rate hikes vs cuts, High vs low inflation).

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
import os; from dotenv import load_dotenv

load_dotenv("../.env")
engine = create_engine(f"postgresql://{os.getenv("POSTGRES_USER")}:{os.getenv("POSTGRES_PASSWORD")}@localhost:5433/risk_db")
```

```
macro_df = pd.read_sql("SELECT * FROM mart_macro_overlay", engine)

# Filter to Tech vs Energy
plot_df = macro_df[macro_df["ticker"].isin(["XLK", "XLE"])]

plt.figure(figsize=(10,6))
sns.barplot(data=plot_df, x="regime_label", y="sharpe_1yr", hue="ticker")
plt.title("Sharpe Ratio by Economic Regime (Tech vs Energy)")
plt.axhline(0, color="black", linestyle="--")
plt.show()
```
