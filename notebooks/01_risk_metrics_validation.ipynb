# 01 Risk Metrics Validation
Validating the Mathematical accuracy of our 95% computing tail risks (VaR).

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
import os
from dotenv import load_dotenv

load_dotenv("../.env")
engine = create_engine(f"postgresql://{os.getenv("POSTGRES_USER")}:{os.getenv("POSTGRES_PASSWORD")}@localhost:5433/risk_db")
```

```
var_df = pd.read_sql("SELECT * FROM mart_var_cvar WHERE ticker = 'SPY'", engine)
var_df["date"] = pd.to_datetime(var_df["date"])

plt.figure(figsize=(12,5))
plt.plot(var_df["date"], var_df["var_95"], color="red", label="95% VaR (Rolling 252d)")
plt.plot(var_df["date"], var_df["var_99"], color="darkred", label="99% VaR")
plt.title("Rolling Value at Risk Profile: SPY")
plt.legend()
plt.show()
```
